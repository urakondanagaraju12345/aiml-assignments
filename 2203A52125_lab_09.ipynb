```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the data
data = pd.read_csv("/content/CC GENERAL.csv")

```

```python
print(data.head())
print(data.info())
```

```python
data.describe()
```

```python
# Check for missing values
print(data.isnull().sum())

# Fill missing values with mean for numeric columns only
for col in data.select_dtypes(include=[np.number]).columns:
    data[col] = data[col].fillna(data[col].mean())



```

```python
print(data.isnull().sum())
```

```python
# Drop the 'CUST_ID' column as it is not useful for clustering
data = data.drop('CUST_ID', axis=1)

# Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

```

```python
# Apply KMeans clustering
kmeans = KMeans(n_clusters=5, random_state=0)  # Change the number of clusters as per your requirement
clusters = kmeans.fit_predict(data_scaled)

# Add the cluster labels to the original data
data['cluster'] = clusters

# Print the first 5 rows of the data with cluster labels
print(data.head())
```
